```python
import os
import tensorflow as tf
from object_detection.utils import label_map_util
from object_detection.utils import visualization_utils as viz_utils
from object_detection.builders import model_builder
from object_detection.utils import config_util
```

```python
# Load pipeline config and build a detection model
configs = config_util.get_configs_from_pipeline_file(files['PIPELINE_CONFIG'])
detection_model = model_builder.build(model_config=configs['model'], is_training=False)

# Restore checkpoint
ckpt = tf.compat.v2.train.Checkpoint(model=detection_model)
ckpt.restore(os.path.join(paths['CHECKPOINT_PATH'], 'ckpt-33')).expect_partial()

@tf.function
def detect_fn(image):
    image, shapes = detection_model.preprocess(image)
    prediction_dict = detection_model.predict(image, shapes)
    detections = detection_model.postprocess(prediction_dict, shapes)
    return detections
```

```python
def filter_text(region, ocr_result, region_threshold):
    rectangle_size = region.shape[0]*region.shape[1]
    
    plate = []
    
    for result in ocr_result:
        length = np.sum(np.subtract(result[0][1], result[0][0]))
        heigth = np.sum(np.subtract(result[0][2], result[0][1]))
        
        if length*height / rectangle_size > region_threshold:
            plate.append(result[1])
    return plate
```

```python
def ocr_it(image, detection, detection_threshold, region_threshold ):
    scores = list(filter (lambda x: x> detection_threshold, detections['detection_scores']))
    boxes = detections['detection_boxes'][:len(scores)]
    classes = detections['detection_classes'][: len(scores)]
    
    width = image.shape[1]
    height = image.shape[0]
    
    # Apply ROI filtering and OCR

    for idx, box in enumerate(boxes): 
        roi = box*[height, width, height, width] 
        region = image[int(roi[0]):int(roi[2]),int(roi[1]):int(roi[3])] 
        reader = easyocr.Reader(['en'])
        ocr_result = reader.readtext(region)
        
        text = filter_text(region, ocr_result, region_threshold)
        
        plt.imshow(cv2.cvtColor(region, cv2.COLOR_BGR2RGB))
        print(text)
        return text, region
    
    
```

```python
import csv
import uuid
import datetime
import os
import cv2
import xlsxwriter
import openpyxl
import re
import json
```

```python
def save_results(text, region, excel_filename, folder_path):
    now = datetime.datetime.now()
    dt_string = now.strftime("%Y-%m-%d_%H-%M-%S")
    img_name = '{}.jpg'.format(dt_string)

    cv2.imwrite(os.path.join(folder_path, img_name), region)

    # Convert text to string
    text_str = str(text)

    # Check if Excel file already exists
    if os.path.isfile(excel_filename):
        # Open the existing Excel file and search for the number plate text
        wb = openpyxl.load_workbook(excel_filename)
        ws = wb.active

        # Remove brackets from the number plate text
        if "(" in text_str and ")" in text_str:
            clean_text = text_str.replace("(", "").replace(")", "")
        else:
            clean_text = text_str

        for row in range(2, ws.max_row + 1):
            if ws.cell(row=row, column=2).value == clean_text:
                # If the number plate text is found, update the existing row with the current date and time
                ws.cell(row=row, column=3, value=now.strftime("%Y-%m-%d"))
                ws.cell(row=row, column=4, value=now.strftime("%H:%M:%S"))

                # Check if the number plate has been detected before
                fifth_column_value = ws.cell(row=row, column=5).value
                if fifth_column_value:
                    try:
                        # If the value in the fifth column is a JSON string, parse it and update the list
                        time_list = json.loads(fifth_column_value)
                    except json.JSONDecodeError:
                        # If the value is not a JSON string, assume it's a single time value and convert it to a list
                        time_list = [fifth_column_value]

                    # Append the current time as string to the list
                    time_list.append(now.strftime("%H:%M:%S"))
                    ws.cell(row=row, column=5, value=json.dumps(time_list))  # Convert list to string
                else:
                    # If the number plate has not been detected before, add the first detection time in the fifth column
                    ws.cell(row=row, column=5, value=json.dumps([now.strftime("%H:%M:%S")]))  # Convert list to string

                # Check the 1st column in the same row for the name
                name = ws.cell(row=row, column=1).value
                if name:
                    print("Updated number plate text '{}' in Excel file".format(clean_text))
                    
                    print("Name: {}".format(name))
                else:
                    print("Updated number plate text '{}' in Excel file".format(clean_text))
                break
        else:
            # If the number plate text is not found, print an error message
            print("Number plate text '{}' not found in Excel file".format(clean_text))
        wb.save(excel_filename)
    else:
        # If the Excel file does not exist, create a new file and add a worksheet
        workbook = openpyxl.Workbook()
        worksheet = workbook.active

        # Remove brackets from the number plate text
        if "(" in text_str and ")" in text_str:
            clean_text = text_str.replace("(", "").replace(")", "")
        else:
            clean_text = text_str

        # Write the data to the worksheet
        row_num = worksheet.max_row + 1
        worksheet.cell(row=row_num, column=1, value=clean_text)
        worksheet.cell(row=row_num, column=2, value=now.strftime("%Y-%m-%d"))
        worksheet.cell(row=row_num, column=3, value=now.strftime("%H:%M:%S"))
        worksheet.cell(row=row_num, column=4, value=json.dumps([now.strftime("%H:%M:%S")]))  # Convert list to string

        workbook.save(excel_filename)
```

# 10. Real Time Detections from your Webcam

```python
cap = cv2.VideoCapture(0)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

while cap.isOpened(): 
    ret, frame = cap.read()
    image_np = np.array(frame)
    
    input_tensor = tf.convert_to_tensor(np.expand_dims(image_np, 0), dtype=tf.float32)
    detections = detect_fn(input_tensor)
    
    num_detections = int(detections.pop('num_detections'))
    detections = {key: value[0, :num_detections].numpy()
                  for key, value in detections.items()}
    detections['num_detections'] = num_detections

    # detection_classes should be ints.
    detections['detection_classes'] = detections['detection_classes'].astype(np.int64)

    label_id_offset = 1
    image_np_with_detections = image_np.copy()

    viz_utils.visualize_boxes_and_labels_on_image_array(
                image_np_with_detections,
                detections['detection_boxes'],
                detections['detection_classes']+label_id_offset,
                detections['detection_scores'],
                category_index,
                use_normalized_coordinates=True,
                max_boxes_to_draw=5,
                min_score_thresh=.8,
                agnostic_mode=False)
    try:
        text, region = ocr_it(image_np_with_detections, detections, detection_threshold, region_threshold )
        save_results(text, region, 'ASi.xlsx', 'det img')
    except:
        pass
    

    cv2.imshow('object detection',  cv2.resize(image_np_with_detections, (800, 600)))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        cap.release()
        cv2.destroyAllWindows()
        break
        
        
        
        
```
